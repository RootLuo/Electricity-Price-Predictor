```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 20,10
```

## Commodity prices: Coal, Natural gas

```python
dfCommodity = pd.read_csv('../data/commodityPrices.csv')
dfCommodity['date'] = pd.to_datetime(dfCommodity['date'])
dfCommodity = dfCommodity.set_index('date').sort_index()
dfCommodity.plot()
dfCommodity.head()
```

```python
dfCommodityDelta = dfCommodity.diff()
dfCommodityDelta.plot()
dfCommodityDelta.head()
```

## Wind Generation

```python
dfWind = pd.read_csv('../data/windGeneration.csv')
dfWind['date'] = pd.to_datetime(dfWind['date'])
dfWind = dfWind.set_index('date').sort_index()
dfWind.plot()
dfWind.head()
```

```python
dfWindDelta = dfWind.diff()
dfWindDelta.plot()
dfWindDelta.head()
```

## MISO Load

```python
dfLoad = pd.read_csv('../data/MISOActualLoad.csv')
dfLoad['Market Day'] = pd.to_datetime(dfLoad['Market Day'])
dfLoad = dfLoad.set_index('Market Day').sort_index()
dfLoad.index.names = ['date']
dfLoad.plot()
dfLoad.head()
```

```python
dfLoadDelta = dfLoad.diff()
dfLoadDelta.plot()
dfLoadDelta.head()
```
